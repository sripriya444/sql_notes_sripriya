#inner join

```python
query="select * from customers c inner join order_details o on c.cust_id = o.order_id"
od3=pd.read_sql(query,engine)
od3
```

```python
query='''select c.cust_name,c.cust_id, count(c.cust_id)
from customers c inner join order_details o on c.cust_id=o.order_id
group by c.cust_id order by count(c.cust_id) desc;'''
od=pd.read_sql(query,engine)
od
```

```python
query='''select c.cust_name,c.cust_id, count(c.cust_id)
from customers c left join order_details o on c.cust_id=o.order_id
group by c.cust_id order by count(c.cust_id) desc;'''
od=pd.read_sql(query,engine)
od
```

```python
query="select * from customers c left join order_details o on c.cust_id = o.order_id"
od=pd.read_sql(query,engine)
od
```

```python
query="select * from customers c right join order_details o on c.cust_id = o.order_id"
od=pd.read_sql(query,engine)
od
```

```python
query="select * from customers c full outer join order_details o on c.cust_id = o.order_id"
od=pd.read_sql(query,engine)
od
```

```python
query="select * from customers c inner join order_details o on c.cust_id = o.order_id and o.qty>2"
od=pd.read_sql(query,engine)
od
```

```python
query="select * from customers c inner join order_details o on c.cust_id = o.order_id and o.item_id in (3,9,7)"
od=pd.read_sql(query,engine)
od
```
